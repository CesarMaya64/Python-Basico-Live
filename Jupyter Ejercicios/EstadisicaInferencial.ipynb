### Correlación

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Crear un conjunto de datos de dos variables
x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])

# Calcular la correlación usando la función corrcoef de NumPy
corr = np.corrcoef(x, y)

# Imprimir el resultado
print("La correlación entre x e y es:", corr[0,1])

# Crear un gráfico de dispersión para visualizar la relación entre las variables
plt.scatter(x, y)
plt.title("Gráfico de dispersión de x e y")
plt.xlabel("Variable x")
plt.ylabel("Variable y")
plt.show()
```

### Análisis Exploratorio de datos (EDA)


```python
import pandas as pd
import seaborn as sns

# Cargar el conjunto de datos de ejemplo "tips" de Seaborn
tips = sns.load_dataset("tips")
```

```python
# Mostrar las primeras filas del conjunto de datos
tips.head()
```

```python
# Resumen estadístico del conjunto de datos
tips.describe()
```

```python
# Histograma de la variable total_bill
sns.histplot(data=tips, x="total_bill", kde=True)
```

```python
# Gráfico de dispersión de las variables total_bill y tip
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# Gráfico de barras de la variable day
sns.countplot(data=tips, x="day")
```
